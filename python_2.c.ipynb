```python
import csv

def print_graduation_eligible_students(student_records):

    student_data = {}

    with open(student_records, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row

        for row in reader:
            roll_number, department, credits, course_name, course_type, grade = row
            credits = int(credits)

            if roll_number not in student_data:
                student_data[roll_number] = {'hasmed_elective': 0, 'honours': 0, 'minor': 0, 'core': 0, 'flexible_elective': 0, 'department_elective': 0, 'alc': 0}

            if course_type in student_data[roll_number]:
                student_data[roll_number][course_type] += credits
            else:
                print(f"Warning: Unexpected course type '{course_type}' for roll number {roll_number}")

        for roll_number, data in student_data.items():
            if (
                data['core'] >= 20 and
                data['department_elective'] >= 15 and
                data['flexible_elective'] >= 10 and
                data['hasmed_elective'] >= 5
            ):
                print(roll_number)

print_graduation_eligible_students("student_records.csv")
```
