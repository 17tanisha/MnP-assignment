```python
import csv

def print_cumulative_cpi(student_records):

    grade_points = {
        'AP': 10,
        'AA': 10,
        'AB': 9,
        'BB': 8,
        'BC': 7,
        'CC': 6
    }

    with open(student_records, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row

        student_data = {}  # Dictionary to store data for each student

        for row in reader:
            roll_no, _, credits, _, _, grade = row
            credits = int(credits)

            if roll_no not in student_data:
                student_data[roll_no] = {'total_credits': 0, 'total_grade_points': 0}

            student_data[roll_no]['total_credits'] += credits
            student_data[roll_no]['total_grade_points'] += credits * grade_points[grade]


        for roll_no, data in student_data.items():
            total_credits = data['total_credits']
            total_grade_points = data['total_grade_points']
            cpi = total_grade_points / total_credits
            print(f"Roll No: {roll_no}, Total Credits: {total_credits}, Cumulative CPI: {cpi:.2f}")

# Example usage:
print_cumulative_cpi("student_records.csv")
```
