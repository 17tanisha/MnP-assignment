```python
import csv

def print_honors_completers(student_records):

    student_data = {}

    with open(student_records, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row

        for row in reader:
            roll_number, department, credits, course_name, course_type, grade = row
            credits = int(credits)

            if roll_number not in student_data:
                student_data[roll_number] = {'honours': 0, 'core': 0}

            if course_type == 'honours':
                student_data[roll_number]['honours'] += credits
            elif course_type == 'core':
                student_data[roll_number]['core'] += credits

        for roll_number, data in student_data.items():
            if data['honours'] >= 10 and data['core'] >= 20:
                print(roll_number)

print_honors_completers("student_records.csv")
```
