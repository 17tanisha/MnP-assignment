```python
import csv

def print_minor_completers(student_records):

    with open(student_records, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row

        student_data = {}

        for row in reader:
            roll_number, department, credits, course_name, course_type, grade = row
            credits = int(credits)

            if roll_number not in student_data:
                student_data[roll_number] = {'minor': 0}

            if course_type == 'minor':
                student_data[roll_number]['minor'] += credits

        for roll_number, data in student_data.items():
            if data['minor'] >= 10:
                print(roll_number)

print_minor_completers("student_records.csv")
```
